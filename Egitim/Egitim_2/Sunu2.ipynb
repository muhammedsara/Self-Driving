```python
# Gerekli kütüphanelerimizi çağırıyoruz
# OpenCv
import cv2
# Numpy
import numpy as np
# Bekleme (sleep) için Zamanla ilgili kütüphane
import time
# Çizim Kütüphanesi
import matplotlib.pylab as plt
%matplotlib inline
```

```python
help(cv2.VideoCapture)
```

```python
# Usb kamera için kullanıldığında parametere olarak sadece kameranın numarasını girmemiz gerekiyor
vcap = cv2.VideoCapture(0)
```

```python
# Görüntü okumak istediğimizde VideoCapture nesnesinin read fonksiyonunu çağırıyoruz
# iki değer döner ilki okumanın başarılı olup olmadığını gösteren değer 
# ikincisi eğer okuma başarılıysa görüntü matrisimiz.
ret,image = vcap.read()
if ret == True:
    print('Image Shape : ',image.shape)
else:
    print('Image Not readed ')
```

```python
for i in range(5):
    ret,image = vcap.read()
    if ret:        
        plt.axis('off')
        #plt.figure(figsize=(20,44))
        plt.imshow(image)
        plt.show()
    time.sleep(1)
```

```python
# Kamerayı kapatmak için release fonksiyonunu çağırıyoruz.
vcap.release()
```

```python
cv2.imwrite('deneme.jpg',image)
```
